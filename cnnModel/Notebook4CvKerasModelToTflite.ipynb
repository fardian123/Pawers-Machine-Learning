```python
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.nn import softmax as softmax_v2
# Load model

```

```python
model = load_model('kucingku_sehat_adamv1.keras', custom_objects={'softmax_v2': softmax_v2})
```

```python
import tensorflow as tf


converter = tf.lite.TFLiteConverter.from_keras_model(model)


converter.optimizations = [tf.lite.Optimize.DEFAULT]

# Tentukan target tipe data float16 (half-precision)
converter.target_spec.supported_types = [tf.float16]

# Konversi model
tflite_model = converter.convert()

# Simpan model TFLite ke file
with open('model_quantized.tflite', 'wb') as f:
    f.write(tflite_model)

print("Model berhasil dikonversi ke format TFLite dengan half-quantization.")
```

```python
# Load TFLite model
interpreter = tf.lite.Interpreter(model_path='model_quantized.tflite')
interpreter.allocate_tensors()

# Periksa detail input dan output
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

print("Input Details:", input_details)
print("Output Details:", output_details)
```
